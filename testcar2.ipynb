# car001

```python
path = r'/content/drive/My Drive/unsup-car-dataset/freiburg_static_cars_52_v1.1/annotations/1_annot.txt'
```

```python
car001 = pd.read_csv(path, delimiter = '\t', names=['all', 'X', 'Angle'],header=None)
car001 
```

```python
car001 = car001.iloc[:,::2]
car001
```

```python
list_car = list(car001['all']) # เก็บที่อยู่ของภาพในรูปแบบตาราง โดยชื่อว่า list_path2 ซึ่ง list(ชื่อตาราง['ชื่อคอลัมน์'])
ID_list2 = list() # สร้างตัวแปร class_list มาเก็บ angle ว่าง
Name_list2 = list()
for i in list_car: # วนลูปใน list_path2
    res = i.split('/') # split ใช้สำหรับแยก โดยในที่นี้แยกด้วย / แล้วตั้งชื่อว่า res
    ID_list2.append(res[0]) # append  คำสั่งเพิ่มข้อมูลลงใน List
    Name_list2.append(res[1])
car001['ID'] = ID_list2 # ตั้งชื่อคอลัมน์ว่า Angle ในตาราง pathname โดยนำค่าใน angle_list2 มาใส่
car001['Name'] = Name_list2
car001 = car001.iloc[:,[2,3,1]]
car001
```

```python
list_car = list(car051['all']) # เก็บที่อยู่ของภาพในรูปแบบตาราง โดยชื่อว่า list_path2 ซึ่ง list(ชื่อตาราง['ชื่อคอลัมน์'])
ID_list2 = list() # สร้างตัวแปร class_list มาเก็บ angle ว่าง
Name_list2 = list()
for i in list_car: # วนลูปใน list_path2
    res = i.split('/') # split ใช้สำหรับแยก โดยในที่นี้แยกด้วย / แล้วตั้งชื่อว่า res
    ID_list2.append(res[0]) # append  คำสั่งเพิ่มข้อมูลลงใน List
    Name_list2.append(res[1])
car051['ID'] = ID_list2 # ตั้งชื่อคอลัมน์ว่า Angle ในตาราง pathname โดยนำค่าใน angle_list2 มาใส่
car051['Name'] = Name_list2
car051 = car051.iloc[:,[2,3,1]]
car051
```
