## test

```python
for i in range(len(car003)):
    ang = car003['Angle'][i]
    if 30 < ang < 46 :
        car003.loc[car003.index[i], 'P'] = 'p08'
    if 45 < ang < 63 :
        car003.loc[car003.index[i], 'P'] = 'p04'
    if 62 < ang < 116 :
        car003.loc[car003.index[i], 'P'] = 'p01,p02'
    if 115 < ang < 136 :
        car003.loc[car003.index[i], 'P'] = 'p03'
    if 135 < ang < 143 :
        car003.loc[car003.index[i], 'P'] = 'p07'
    if 142 < ang < 181 :
        car003.loc[car003.index[i], 'P'] = 'p05'
    if 180 < ang < 226 :
        car003.loc[car003.index[i], 'P'] = 'p09'
    if 225 < ang < 251 :
        car003.loc[car003.index[i], 'P'] = 'p11'
    if 250 < ang < 291 :
        car003.loc[car003.index[i], 'P'] = 'p13, p14'   ## if ',' in car052['p'] , split(',')  
    if 290 < ang < 311 :
        car003.loc[car003.index[i], 'P'] = 'p12' 
    if 310 < ang < 358 :
        car003.loc[car003.index[i], 'P'] = 'p10'
    elif 357 < ang < 361 or ang <= 30 :
        car003.loc[car003.index[i], 'P'] = 'p06'     
```
